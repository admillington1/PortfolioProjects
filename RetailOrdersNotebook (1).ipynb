```python
#read data and handle null values
import pandas as pd
df = pd.read_csv('orders.csv', na_values = ['Not Available', 'unknown'])
```

```python
df.head(20)
```

```python
df['Ship Mode'].unique()
```

```python
df.columns
```

```python
#rename columns to lower case and replace space with underscore
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df.columns
```

```python
df.head()
```

```python
#create new columns for the discount amount, sale price and the profit
df['discount_amount'] = df['list_price'] * df['discount_percent'] * 0.01
df['sale_price'] = df['list_price'] - df['discount_amount']
df['profit'] = df['sale_price'] - df['cost_price']
df
```

```python
#find data types
df.dtypes
```

```python
#order_date is an object and we need to convert it to datetime
pd.to_datetime(df['order_date'], format = '%Y-%m-%d')
```

```python
#We don't need cost_price, discount_percentage or list_price so we will drop them
df.drop(columns = ['cost_price', 'list_price', 'discount_percent'], inplace = True)
```
